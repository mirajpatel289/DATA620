### Dataset: Florentine Families Network
This notebook explores the social structure of 15th-century Italy using the **Florentine Families** dataset.

```python
import networkx as nx
import matplotlib.pyplot as plt

#Load the built-in Florentine Families graph
G = nx.florentine_families_graph()

#Calculate Diameter (1st metric)
diameter = nx.diameter(G)

#Calculate Betweenness Centrality (2nd metric)
centrality = nx.betweenness_centrality(G)

print(f"Graph Diameter: {diameter}")
print(f"Top Centrality Score: {max(centrality.values()):.4f}")
```

```python
#Export to GEXF format
nx.write_gexf(G, "florentine_graph.gexf")
```
