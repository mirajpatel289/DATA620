```python
import networkx as nx
import matplotlib.pyplot as plt

#Create an undirected graph
G = nx.Graph()

#Add edges based on the Krackhardt Kite layout
edges = [
    ('Andre', 'Beverly'), ('Andre', 'Carol'), ('Andre', 'Diane'), ('Andre', 'Fernando'),
    ('Beverly', 'Diane'), ('Beverly', 'Garth'), ('Beverly', 'Ed'),
    ('Carol', 'Diane'), ('Carol', 'Fernando'),
    ('Diane', 'Ed'), ('Diane', 'Garth'), ('Diane', 'Fernando'),
    ('Ed', 'Garth'),
    ('Fernando', 'Garth'), ('Fernando', 'Heather'),
    ('Garth', 'Heather'),
    ('Heather', 'Ike'),
    ('Ike', 'Jane')
]

G.add_edges_from(edges)

#Draw the graph
plt.figure(figsize=(8, 5))
nx.draw(G, with_labels=True, node_color='lightblue', edge_color='purple', node_size=2000, font_weight='bold')
plt.show()
```
